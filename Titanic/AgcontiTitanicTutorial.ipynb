```python
# Let's read the data
import pandas as pd

df = pd.read_csv("train.csv")

df
```

```python
# Drop Ticket and Cabin features, because they have too many missing values
df = df.drop(["Ticket", "Cabin"], axis=1)

# Drop NaN values
df = df.dropna()
```

```python
# Make graphs of the data
import matplotlib.pyplot as plt
%matplotlib inline

# Specifices the parameters of our graphs
fig = plt.figure(figsize=(18,6), dpi=1600)
alpha_scatterplot = 0.2
alpha_bar_chart = 0.55

# To plot many different graphs together
ax1 = plt.subplot2grid((2,3), (0,0))

# Plots a bar graph of those who survived and those who did not
df.Survived.value_counts().plot(kind="bar", alpha=alpha_bar_chart)

# Sets graph horizontal range
ax1.set_xlim(0, 2)

# Puts a title
plt.title("Distribution of Survived, (1 = Survived)")

plt.subplot2grid((2,3), (0,1))
plt.scatter(df.Survived, df.Age, alpha=alpha_scatterplot)

# Sets the y axis label
plt.ylabel("Age")

# Formats the grid line style
plt.grid(b=True, which="major", axis='y')
plt.title("Survived by Age, (1 = Survived)")

ax3 = plt.subplot2grid((2,3), (0,2))
df.Pclass.value_counts().plot(kind="barh", alpha=alpha_bar_chart)
ax3.set_ylim(-1, len(df.Pclass.value_counts()))
plt.title("Pclass distribution")

# Plots a kernel of density estimate of the subest of the 1st class passenger's age
plt.subplot2grid((2,3), (1,0), colspan=2)
df.Age[df.Pclass == 1].plot(kind="kde")
df.Age[df.Pclass == 2].plot(kind="kde")
df.Age[df.Pclass == 3].plot(kind="kde")
plt.xlabel("Age")
plt.title("Age distribution within Pclasses")
# Sets legend
plt.legend(("Pclass = 1", "Pclass = 2", "Pclass = 3"), loc="best")

ax5 = plt.subplot2grid((2,3), (1,2))
df.Embarked.value_counts().plot(kind="bar", alpha=alpha_bar_chart)
ax5.set_xlim(-1, len(df.Embarked.value_counts()))
plt.title("Passengers per boarding location")

plt.show()
```
